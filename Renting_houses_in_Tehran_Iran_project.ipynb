# Dealing with region column

```python
# strip removes starting and ending spaces
df['region']=df['region'].apply(lambda x: x.strip())
```

```python
len(df.region.unique())
```

```python
region_stat=df.groupby('region')['region'].agg('count').sort_values(ascending=False)
region_stat
```

```python
region_less_than_100 = region_stat[region_stat<=100]
len(region_less_than_100)
```

```python
df.region=df.region.apply(lambda x: 'other' if x in region_less_than_100 else x)
len(df.region.unique())
```

```python
df[df['region']=='آباد']
```

```python
unique_regions=df.region.unique()
unique_regions[20]
```

```python
region_translated={}
count=0
for i in unique_regions:
    region_translated[i]="R"+str(count)
    count+= 1

```

```python
region_translated
```

```python
print(list(region_translated.keys())[list(region_translated.values()).index('R189')]) 
```

```python
df['new_region']=df['region'].map(region_translated)
df.head()
```

```python
dummies=pd.get_dummies(df.new_region,drop_first=True)
dummies.head()
```

```python
len(df.region.unique())
```

```python
df['price_per_meter']=df.price/df['size(m)']
df.head()
```

```python
df=pd.concat([df,dummies],axis='columns')
#df=df.drop(['region'],axis='columns')
df.head()
```

```python
def t_to_time(timestamp):
    return datetime.fromtimestamp(timestamp)

df['time']=df.apply(lambda row: t_to_time(row['time']), axis=1)
```

```python
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['time']).year
df.head()
```

```python
len(df.year.unique())
```

```python
y=df.price
y
```

```python
X=df.drop(['deposit','rent','year','price','Unnamed: 0','time','region','new_region'],axis=1)
X
```

# Scaling the features

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
X_scaled=scale.fit_transform(X)
```

# sklearn.model_selection.train_test_split


```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(X_scaled,df.price, test_size=0.2)
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=10)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import metrics
print ("r-square= ", metrics.r2_score(y_test, y_pred))
```
